# 16. pandas 合并 concat

```python
import numpy as np
import pandas as pd
```

## concatnating

```python
df1 = pd.DataFrame(np.ones((3,4)) * 0, columns=['a','b','c','d'])
df2 = pd.DataFrame(np.ones((3,4)) * 1, columns=['a','b','c','d'])
df3 = pd.DataFrame(np.ones((3,4)) * 2, columns=['a','b','c','d'])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1, df2, df3], axis=0, ignore_index=True)
```

```python
pd.concat([df1, df2, df3], axis=1, ignore_index=True)
```

## join :  ['inner', 'outer']

```python
df4 = pd.DataFrame(np.ones((3,4)) * 1, columns=['a','b','c','d'], index=[1, 2, 3])
df5 = pd.DataFrame(np.ones((3,4)) * 2, columns=['b','c','d','e'], index=[2, 3, 4])
```

```python
pd.concat([df4, df5], axis=0, ignore_index=True, sort=False)
```

```python
pd.concat([df4, df5], join='outer')   # join 的处理模式默认为 outer
```

```python
pd.concat([df4, df5], join='inner', ignore_index=True) 
```
