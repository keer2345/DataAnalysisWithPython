# 14. pandas处理丢失是数据

```python
import numpy as np
import pandas as pd

dates = pd.date_range('20181201',periods=6)
df = pd.DataFrame(np.arange(24).reshape((6, 4)), index = dates, columns=['A','B','C','D'])
df
```

```python
df.iloc[0, 1] = np.nan
df.iloc[1, 2] = np.nan
df
```

```python
df.dropna(axis=0, how='any')  # how = {'any', 'all'}   当为 all 时，表示所有改行所有值为 NaN 才丢弃
```

```python
df.fillna(value='999')
```

```python
df.isnull()
```

```python
np.any(df.isnull())
```
