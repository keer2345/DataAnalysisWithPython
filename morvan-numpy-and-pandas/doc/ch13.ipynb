# 13. pandas设置值

```python
import numpy as np
import pandas as pd
```

```python
dates = pd.date_range('20181201', periods=6)
df = pd.DataFrame(np.arange(24).reshape((6,4)), index = dates, columns=['A','B','C','D'])
df
```

```python
df.iloc[2,3]=111
df
```

```python
df.loc['20181203','A'] = 222
df
```

```python
df[ df.A > 4 ] = 0
df
```

```python
df['E'] = np.nan
df
```

```python
df['F'] = pd.Series([1,2,3,4,5,6], index =pd.date_range('20181201', periods=6))
df
```
