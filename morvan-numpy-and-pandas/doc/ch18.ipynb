# 18. pandas plot

```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

## Series

```python
data = pd.Series(np.random.randn(1000), index=np.arange(1000))
data
```

```python
data.cumsum()  
```

```python
data.cumsum().plot()
plt.show()
```

```python
data2 = pd.DataFrame(np.random.randn(1000,4), index= np.arange(1000), columns=list("ABCD"))
data2
```

```python
data2 = data2.cumsum()
data2.plot()
plt.show()
```
