# 17. pandas 合并 merge

## merge

```python
import numpy as np
import pandas as pd
```

```python
left = pd.DataFrame({'key':['k0','k1','k2','k3'],
                    'a':['a0','a1','a2','a3'],
                    'b':['b0','b1','b2','b3']})
left
```

```python
right = pd.DataFrame({'key':['k0','k1','k2','k3'],
                    'c':['c0','c1','c2','c3'],
                    'd':['d0','d1','d2','bd3']})
right
```

```python
res = pd.merge(left, right,on='key')
```

```python
res
```

```python
left2 = pd.DataFrame({'key1':['k0','k0','k1','k2'],
                      'key2':['k0','k1','k0','k1'],
                    'a':['a0','a1','a2','a3'],
                    'b':['b0','b1','b2','b3']})
left2
```

```python
right2 = pd.DataFrame({'key1':['k0','k1','k1','k2'],
                       'key2':['k0','k0','k0','k0'],
                    'c':['c0','c1','c2','c3'],
                    'd':['d0','d1','d2','bd3']})
right2
```

```python
res2 = pd.merge(left2, right2, on = ['key1', 'key2'])
res2
```

```python
res3 = pd.merge(left2, right2, on = ['key1', 'key2'], how='outer')   # how 默认为　inner
res3
```

```python
res4 = pd.merge(left2, right2, on = ['key1', 'key2'], how='right')  
res4
```

```python
res5 = pd.merge(left2, right2, on = ['key1', 'key2'], how='left')  
res5
```

```python
res6 = pd.merge(left2, right2, on = ['key1', 'key2'], how='outer', indicator=True) 
res6
```

```python
res7 = pd.merge(left2, right2, on = ['key1', 'key2'], how='outer', indicator='indicator_column') 
res7
```

## merged by index

```python
left3 =  pd.DataFrame({'a':['a0','a1','a2'],
                      'b':['b0','b1','b2']},
                     index=['k0','k1','k2'])
left3
```

```python
right3 =  pd.DataFrame({'c':['c0','c2','c3'],
                      'd':['d0','d2','d3']},
                     index=['k0','k2','k3'])
right3
```

```python
res8 = pd.merge(left3, right3, left_index=True, right_index=True, how='inner')
res8
```

```python
res9 = pd.merge(left3, right3, left_index=True, right_index=True, how='outer')
res9
```

```python
boys = pd.DataFrame({'k':['k0','k1','k2'], 'age':[1, 2, 3]})
boys
```

```python
girls = pd.DataFrame({'k':['k0','k0','k3'], 'age':[4, 5, 6]})
girls
```

```python
res10 = pd.merge(boys, girls, on='k', suffixes=['_boy','_girl'], how='inner')
res10
```

```python
res11 = pd.merge(boys, girls, on='k', suffixes=['_boy','_girl'], how='outer')
res11
```
