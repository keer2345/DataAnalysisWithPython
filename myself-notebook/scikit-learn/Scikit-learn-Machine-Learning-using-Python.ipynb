# 什么是机器学习
机器学习是人工智能的一种,它允许应用软件从数据中学习并更加准确的预测结果而无需人工干预。但是怎么实现的？为此，机器需要一些数据并基于数据的训练，它讲探测模式并创建模型。从数据中获取知识并提供强大见解的过程都是与机器学习相关。可以参考以下图片来更好地理解它的工作：

![](https://d1jnx9ba8s6j9r.cloudfront.net/blog/wp-content/uploads/2017/07/MachineLearning-Scikit-Learn-Edureka.jpg)

使用数据，系统学习算法，然后用它来构建预测模型。之后，我们根据反馈数据调整模型或提高模型精度。使用反馈数据，我们在新的数据集上优化模型预测行为。我们讨论算法的一个用例，训练和测试数据有助于我们更好的了解 Scikit learn 是否适合特定问题的解决。

下面有三种类型的机器学习：
1. **监督学习**：(Supervised learning) 这是一个过程，一个算法学习训练数据集。监督学习是在数据变量（X）和输出变量（Y）间使用算法生成映射函数，因为使用数据来做预测的过程，也被称为预测模型。算法包含线性回归(Linear Regression)，逻辑回归(Logistic Regression)，决策树(Decision tree)，随机森林(Random forest) 和朴素贝叶斯(Native Bayes)分类器。我们将进一步讨论监督学习的用例，训练机器使用逻辑回归。
1. **无监督学习**：(Unsupervised learning) 这是一个模型通过无标签的信息来训练的过程。该过程可用于集群输入数据类的基础上统计它们的特性。无监督学习也被成为聚类分析，意味着基于信息的分组对象中发现数据描述或者关系。目标是这些对象在一组中应该相似但有别于其他组。这些算法包括K均值聚类(K-means clustering)，层次聚类(Hierarchical clustering)等。
1. **强化学习**：(Reinforcement learning) 强化学习是与空间或环境交互的学习，强化代理学习可以从其行为后果，而不是显示地接受教育。它依据过去的经验（剥削）来选择行为并做进一步的选择（探索）。

我们也可以形象化这目标标签成图像，看以下代码：

````python
# join the images and target labels in a list
images_and_labels = list(zip(digits.images, digits.target))

# for every element in the list
for index, (image, label) in enumerate(images_and_labels[:8]):
    # initialize a subplot of 2x4 at the i+1-th position
    plt.subplot(2, 4,index+1)
    # display iamges in all subplots
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    # add a title to each subplot
    plt.title('Training: ' + str(label))
    
# show the plot
plt.show()
````

可以看到，我们使用 *zip* 函数来在列表中连接图像和目标标签，然后保存为一个变量，叫做 `images_and_labels` 。之后，我们将网格中的前八个元素（2x4）进行索引，并通过 Matplotlib 来展示图像，设置标题为 "traning"。
